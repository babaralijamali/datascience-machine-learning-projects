# Booking.com Reviews Visualization

```python
import numpy as np # linear algebra
import matplotlib.pyplot as plt 
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)

hotel_reviews = pd.read_csv('hotelr.csv',parse_dates=['Review_Date'])
hotel_reviews['Review_Month']=hotel_reviews['Review_Date'].dt.month
hotel_reviews['Review_Year']=hotel_reviews['Review_Date'].dt.year
```

```python
hotel_reviews.head(5)
```

```python
#Avreage Rating For Hotels
Rating_Frame_1 = hotel_reviews.groupby(['Hotel_Name','Average_Score'], as_index=False).mean('Average_Score')
Rating_Frame_1.sort_values('Average_Score',ascending=False)
Rating_Frame_1
```

```python
#Ratings by Nationalities
Rating_Frame_2 = hotel_reviews.groupby(['Reviewer_Nationality'], as_index=False).mean('Average_Score')
Rating_Frame_2.sort_values('Average_Score',ascending=False)
```

```python
#Top Raters by Country on Booking.com
xAx=Rating_Frame_2.sort_values('Average_Score',ascending=False)['Reviewer_Nationality'].head(10)
yAx=Rating_Frame_2.sort_values('Average_Score',ascending=False)['Average_Score'].head(10)
plt.figure(figsize=(17,8))
plt.bar(xAx,yAx,color='r')
plt.xticks(rotation = 45)
x=0.0
for xItem, yItem in zip(xAx,yAx):
    
    plt.text(x-0.1, 8, '({})'.format(round(yItem,2)),fontsize=12,
             verticalalignment='center')
    x=x+1
```

```python
#Reviwers by Country who Give lowest Ratings
xAx=Rating_Frame_2.sort_values('Average_Score',ascending=True)['Reviewer_Nationality'].head(10)
yAx=Rating_Frame_2.sort_values('Average_Score',ascending=True)['Average_Score'].head(10)
plt.figure(figsize=(17,8))
plt.bar(xAx,yAx,color='g')
plt.xticks(rotation = 45)
x=0.0
for xItem, yItem in zip(xAx,yAx):
    
    plt.text(x-0.2,yItem+0.22, '({})'.format(round(yItem,2)),fontsize=12,
             verticalalignment='center')
    x=x+1
```

```python
# UK Hotels Data
#Assuming below dataframe returns all hotels in UK
UK_Frame=hotel_reviews[hotel_reviews['Hotel_Address'].str.contains('UNITED',case=False)]
UK_Frame
```

```python
#2017 Hotel Ratings UK
Ratings_2017 = UK_Frame[UK_Frame['Review_Year']==2017].groupby(['Hotel_Name']).mean('Average_Score')
Ratings_2017.sort_values('Average_Score', ascending=False)
```

```python
#Top and bottom Most Hotels in 2017, in UK by Rating

top_10_UK = Ratings_2017.sort_values('Average_Score', ascending=False)['Average_Score'].head(10)
bottom_10_UK = Ratings_2017.sort_values('Average_Score', ascending=True)['Average_Score'].head(10)

bottom_10_UK , top_10_UK
```

```python
#Qualifying the average ratings of hotels into categories for 2017
Ratings_2017['Rating'] = Ratings_2017['Average_Score'].apply(lambda x: 'Excellent' if x > 8.5 
                                                             else( 'Very Good' if (x >= 8 and x >= 8.5)
                                                            else ('Good' if (x >= 7 and x < 8)
                                                             else ('Average' if x >= 5 and x < 7
                                                            else 'Bad'))))
Ratings_2017
```

```python
#Charts for average rating categories for 2017
Rating_count = Ratings_2017['Rating'].value_counts()

plt.hist(Ratings_2017['Rating'],edgecolor='black',color='orange', align='mid', linewidth=0.5)    
plt.grid(True)
plt.show()
```

```python
plt.pie(Rating_count,labels=Rating_count.index.values.tolist(),radius=2.1,textprops={'fontsize': 14}, shadow=True)
```
